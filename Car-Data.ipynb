# Creating table called car with colum names as in given data

```python
query = "create table cardataset.car(buying varchar(20), maint varchar(20), doors varchar(20), persons varchar(20), luggage varchar(20), safety varchar(20), class varchar(20))"
cursor.execute(query)
```

```python
query = "create table cardataset.car1(buying varchar(20), maint varchar(20), doors varchar(20), persons varchar(20), luggage varchar(20), safety varchar(20), class varchar(20))"
cursor.execute(query)
```

```python
#Reading data
file = open("car.data",'r')
file.read()
```

```python
#Method-1 to insert data in the table
import csv
with open('car.data','r') as f:
    car_data = csv.reader(f, delimiter='\n')
    for i in car_data:
        row1 = tuple(i[0].split(','))
        query = "INSERT INTO cardataset.car VALUES {}".format(row1)
        cursor.execute(query)
```

```python
#Method-2 to insert data in table
import csv
with open('car.data','r') as f:
    car_data = csv.reader(f, delimiter='\n')
    for i in car_data:
        row1 = i[0].split(',')
        query = "INSERT INTO cardataset.car1 VALUES ('{}','{}','{}','{}','{}','{}','{}')".format(str(row1[0]), str(row1[1]), str(row1[2]), str(row1[3]), str(row1[4]), str(row1[5]), str(row1[6]))
        cursor.execute(query)
```
